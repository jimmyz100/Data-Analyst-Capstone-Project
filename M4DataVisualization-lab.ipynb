## Demo: How to work with database


Download database file.


```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
QUERY = """
SELECT ConvertedComp
FROM master
"""
ConvCompData = pd.read_sql_query(QUERY,conn)
histo = sns.histplot(data=ConvCompData)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
AgeData = pd.read_sql_query(QUERY,conn)
Box = sns.boxplot(data=AgeData)
Box.set_ylabel("Age (years)")
Box.set(title='Box Plot Distribution of Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
ScatterData = pd.read_sql_query(QUERY,conn)
Scatter = sns.scatterplot(x="Age", y="WorkWeekHrs", data=ScatterData)
Scatter.set(title='Age vs WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
BubbleData = pd.read_sql_query(QUERY,conn)
sns.set(rc = {'figure.figsize':(15,8)})
Bubble = sns.scatterplot(x="WorkWeekHrs", y="CodeRevHrs", data=BubbleData, size='Age', legend=False, sizes=(20, 2000))
Bubble.set(title='WorkWeekHrs vs CodeRevHrs Bubble Plot by Age')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*)
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear) DESC
LIMIT 5
"""
PieData = pd.read_sql_query(QUERY,conn)
import matplotlib.pyplot as plt
plt.pie(PieData["COUNT(*)"].tolist(), labels = PieData["DatabaseDesireNextYear"].tolist(),autopct='%.0f%%')
plt.title('Top 5 Desired Databases to Learn Next Year')
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*)
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY COUNT(LanguageDesireNextYear) DESC
LIMIT 5
"""
LangData = pd.read_sql_query(QUERY,conn)
LangData
```

```python
QUERY = """
SELECT LanguageWorkedWith, COUNT(*)
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
"""
LangCurData = pd.read_sql_query(QUERY,conn)
LangCurData
```

```python
QUERY = """
SELECT Respondent, DatabaseWorkedWith, COUNT(*)
FROM DatabaseWorkedWith
GROUP BY Respondent HAVING COUNT(DatabaseWorkedWith)=1 AND DatabaseWorkedWith='MySQL'
"""
CurData = pd.read_sql_query(QUERY,conn)
len(CurData)
```

```python
QUERY = """
SELECT DatabaseWorkedWith, COUNT(*)
FROM DatabaseWorkedWith
GROUP BY DatabaseWorkedWith
"""
DataCurData = pd.read_sql_query(QUERY,conn)
DataCurData
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
StackedData = pd.read_sql_query(QUERY,conn)
StackedData_Age = StackedData.groupby('Age',axis=0)
Age_Median = StackedData_Age.median()

Age_Median.plot(kind='bar', stacked=True)
plt.title('Median Work Hrs by Age')
plt.xlabel('Age')
plt.ylabel('Work Hours')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
LineData = pd.read_sql_query(QUERY,conn)
LineData_Age = LineData.groupby('Age',axis=0)
LineAge_Median = LineData_Age.median()

Line = sns.lineplot(data=LineAge_Median, x="Age", y="ConvertedComp")
Line.set(title='Median ConvertedComp by Age')
```

```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
"""
Line2Data = pd.read_sql_query(QUERY,conn)
Line2Data_Age = Line2Data.groupby('Age',axis=0)
Line2Age_Median = Line2Data_Age.median()

Line2 = sns.lineplot(data=Line2Age_Median, x="Age", y="ConvertedComp")
Line2.set(title='Median ConvertedComp by Age')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as Number_of_Respondents
FROM master
GROUP BY MainBranch
"""
BarHData = pd.read_sql_query(QUERY,conn)
BarH = sns.barplot(x="Number_of_Respondents", y="MainBranch", data=BarHData, color="b")
BarH.set(title='Respondents by Main Branch')
```

```python
QUERY = """
SELECT DevType, COUNT(*)
FROM DevType
GROUP BY DevType
ORDER BY COUNT(DevType) DESC
"""
DevData = pd.read_sql_query(QUERY,conn)
DevData
```

```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
"""
Line1Data = pd.read_sql_query(QUERY,conn)
Line1Data_Age = Line1Data.groupby('Age',axis=0)
Line1Age_Median = Line1Data_Age.median()

Line1 = sns.lineplot(data=Line1Age_Median, x="Age", y="ConvertedComp")
Line1.set(title='Median ConvertedComp by Age')
```

```python
QUERY = """
SELECT LanguageWorkedWith, COUNT(*) as Number_of_Respondents
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY COUNT(LanguageWorkedWith) DESC
LIMIT 5
"""
import seaborn as sns
Top5LangData = pd.read_sql_query(QUERY,conn)
Top5Lang = sns.barplot(x="Number_of_Respondents", y="LanguageWorkedWith", data=Top5LangData, color="b")
Top5Lang.set(title='Top 5 Programming Languages')
```

```python
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as Number_of_Respondents
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY COUNT(LanguageDesireNextYear) DESC
LIMIT 5
"""

Top5LangDesireData = pd.read_sql_query(QUERY,conn)
Top5LangDesire = sns.barplot(x="Number_of_Respondents", y="LanguageDesireNextYear", data=Top5LangDesireData, color="r")
Top5LangDesire.set(title='Top 5 Programming Languages to Learn Next Year')
```

```python
QUERY = """
SELECT DatabaseWorkedWith, COUNT(*) as Number_of_Respondents
FROM DatabaseWorkedWith
GROUP BY DatabaseWorkedWith
ORDER BY COUNT(DatabaseWorkedWith) DESC
LIMIT 5
"""

Top5DataData = pd.read_sql_query(QUERY,conn)
Top5Data = sns.barplot(x="Number_of_Respondents", y="DatabaseWorkedWith", data=Top5DataData, color="g")
Top5Data.set(title='Top 5 Databases')
```

```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Number_of_Respondents
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear) DESC
LIMIT 5
"""

Top5DataDesireData = pd.read_sql_query(QUERY,conn)
Top5DataDesire = sns.barplot(x="Number_of_Respondents", y="DatabaseDesireNextYear", data=Top5DataDesireData, color="y")
Top5DataDesire.set(title='Top 5 Databases to Learn Next Year')
```
